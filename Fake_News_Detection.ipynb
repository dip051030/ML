```python
news_df = news_df.fillna('')
```

```python
news_df.isna().sum()
```

```python
news_df['content'] = news_df['title'] + ' ' + news_df['author']
```

```python
news_df
```

```python
X = news_df.drop('label', axis=1)
y = news_df['label']
```

```python
X.shape 
```

```python
y.shape
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content= re.sub('[^a-zA-Z]', ' ', content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if word not in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
news_df['content'] = news_df.apply(lambda x: stemming(x['content']), axis=1)
```

```python
news_df
```

```python
X = news_df['content'].values
y = news_df['label'].values
```

```python
y
```

```python
#since machine cant understand the textual data, i will transform into tfidfmatrix using tfidfvectoriser.

vectoriser = TfidfVectorizer()
vectoriser.fit(X)
X = vectoriser.fit_transform(X)
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model  = LogisticRegression(C=10)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(model.predict(X_train), y_train))
print(accuracy_score(model.predict(X_test), y_test))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(LogisticRegression(), X_train, y_train, cv=5)
print("Mean CV Accuracy:", scores.mean())
```
